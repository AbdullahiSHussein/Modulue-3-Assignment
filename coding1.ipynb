# Importing libraries 

```python
#import required libraries

from bs4 import BeautifulSoup
import requests
import random
import pandas as pd
```

## Reterving data from wikipedia 

```python
#check we can get the data from the page

#here we use 'query' for the end of the url, this allows us to quickly change it
query = 'AFC_Wimbledon'

url = 'https://en.wikipedia.org/wiki/' + query
page = requests.get(url)
soup = BeautifulSoup(page.content, features="html.parser")
print(soup)

#this grabs us the html of the entire page
```

```python
# get the response in the form of html
wikiurl="https://en.wikipedia.org/wiki/Football"

# check the request was sucsessful (code 200)
response=requests.get(wikiurl)
print(response.status_code)

# parse data from the html into a beautifulsoup object
soup = BeautifulSoup(response.text, 'html.parser')

# here we find any element with the table tag, there are some of these we dont want on this page. 
# So we specify only tables using the "wikitable" class
 
if page.status_code == 200:
    # Find all 'p' tags and extract text
    text = []
    for p_tag in soup.find_all('p'):
        text.append(p_tag.get_text())

    # Join all the text into a single string and print
    article_text = "\n".join(text)
    print(article_text)
else:
    print(f"Failed to retrieve the page. Status code: {page.status_code}")
```

## Text to numbers 

```python
import numpy as np
from collections import Counter
```

```python
#tokens is a 1D array
tokens = article_text.split() 
#Get the unique tokens (our vocabulary)
vocab = np.unique(tokens)
print("total words:",len(tokens), "unique words:", len(vocab))
#Create a Bag of Words using a Counter
Counter(tokens).most_common(200)
```

```python
# find duplicate words that might be taken 
import re 
ctr = 0; # counter
```

```python
for word in vocab: 
    if len(re.findall(r'[\.,;!?]$', word)) > 0: # `re.findall(r'[\.,;!?]$', word)` returns an empty list if the word does not end ($) in . , ; ! or ?.
        if word[:-1] in vocab:
            ctr = ctr + 1 # this is also commonly written as `ctr +=1`
print(ctr," duplicate words ending in [.,;!?] out of ", len(vocab))
```

## Removing punctuation duplicates 

```python

#Use a regex to split based on space AND punctuation
tokens = re.split(r'[-\s.,;!?()]+', article_text) 
vocab = np.unique(tokens)
print("unique words", vocab.shape)
Counter(tokens).most_common(50)
```

## Stemming

```python
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()
#Looking at our hacking book
stem_tokens = []
for t in tokens:
    stem_tokens.append(stemmer.stem(t))
stem_vocab = np.unique(stem_tokens)
print("unique words", stem_vocab.shape)
Counter(stem_tokens).most_common(50)
```

## Lemmatisation

```python
import nltk
nltk.download("wordnet")
nltk.download('omw-1.4')
from nltk.stem import WordNetLemmatizer
lem = WordNetLemmatizer()
```

```python

#Looking at our hacking book
lem_tokens = []
for t in tokens:
    lem_tokens.append(lem.lemmatize(t))
lem_vocab = np.unique(lem_tokens)
print("unique words", lem_vocab.shape)
Counter(lem_tokens).most_common(50)
```

## Capitalisation

```python
lower_tokens = []
for word in lem_tokens:
    lower_tokens.append(word.lower())
lower_vocab = np.unique(lower_tokens)
print("unique words", lower_vocab.shape)
bow = Counter(lower_tokens)
bow.most_common(50)
```

# Generative text with markopv chain and rnns 

## Create a Word-Based Transition Matrix

```python
import random
from collections import defaultdict

clearn_text = " " .join (tokens) 

# Split the text into words
words = clearn_text.split()

# Create a dictionary to store transitions
transition_matrix = defaultdict(list)

# Populate the transition matrix with word pairs
for i in range(len(words) - 1):
    current_word = words[i]
    next_word = words[i + 1]
    transition_matrix[current_word].append(next_word)


print(clearn_text)
```

```python
def generate_sentence(start_word, length):
    current_word = start_word
    sentence = [current_word]

    for _ in range(length):
        next_word = random.choice(transition_matrix[current_word])
        sentence.append(next_word)
        current_word = next_word

    return " ".join(sentence)

# Generate a sentence starting with the word "Alice"
print(generate_sentence("football", 20))
```
